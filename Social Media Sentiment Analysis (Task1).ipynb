```python
#we select the necessary columns for creating the support vector machine 

df = df[['text','airline_sentiment']]
```

```python
#we split the data into text and labels

X = df['text']
Y = df['airline_sentiment']
```

```python
#we vectorize the text data for the LinearSVC

vectorizer = TfidfVectorizer()
X_vectors = vectorizer.fit_transform(X)
```

```python
#we now train the data

classifier = LinearSVC()
classifier.fit(X_vectors, Y)
```

```python
#now, we predict the required data for the task,i.e the sentiments

Y_pred = classifier.predict(X_vectors)
df['predicted_sentiment'] = Y_pred
```
